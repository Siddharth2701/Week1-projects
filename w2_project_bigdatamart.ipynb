```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt 
import seaborn as sb
```

```python
df=pd.read_csv('bigdatamart_train.csv')
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df1=df.drop(columns=['Item_Weight'])
df1.head()
```

```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy='most_frequent')
df1['Outlet_Size']=imp.fit_transform(df1['Outlet_Size'].values.reshape(-1,1))
```

```python
df1
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
list1=['Item_Identifier','Item_Fat_Content','Item_Type','Outlet_Identifier','Outlet_Size','Outlet_Location_Type','Outlet_Type']
for val in list1:
    df1[val]=le.fit_transform(df[val].astype(str))

df1
```

```python
df1.isnull().sum()
```

```python
df1.describe()
```

```python
x=df1.drop(columns='Item_Outlet_Sales',axis=1)
x
```

```python
y=pd.DataFrame(df1['Item_Outlet_Sales'])
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=47,test_size=.22)
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
pred=reg.predict(x_test)
print(r2_score(y_test,pred))
```

```python
print(mean_squared_error(y_test,pred))
```

```python
print(mean_absolute_error(y_test,pred))
```
